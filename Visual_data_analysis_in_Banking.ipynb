```python
age_df=(dataf.groupby('age')[["y"]].sum().join(dataf.groupby('age')[['y']].count(),rsuffix='_count'))
age_df.columns=["Attracted","Total_number"]
```

```python
trace0=go.Scatter(x=age_df.index,y=age_df["Attracted"],name="Attracted")
trace1=go.Scatter(x=age_df.index,y=age_df["Total_number"],name="Total_number")
data=[trace0,trace1]
layout={"title":"Statistics by client age"}
fig=go.Figure(data=data,layout=layout)
iplot(fig,show_link=False)
```

```python
#Lets see the customers by the months
```

```python
dataf.head()
```

```python
month_index = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
month_df = (
    dataf.groupby("month")[["y"]]
    .sum()
    .join(dataf.groupby("month")[["y"]].count(), rsuffix='_count')
).reindex(month_index)
month_df.columns = ["Attracted", "Total Number"]
```

```python
trace0 = go.Bar(x=month_df.index, y=month_df["Attracted"], name="Attracted")
trace1 = go.Bar(x=month_df.index, y=month_df["Total Number"], name="Total Number")

data = [trace0, trace1]
layout = {"title": "Share of months"}

fig = go.Figure(data=data, layout=layout)

iplot(fig, show_link=False)
```

```python
cate=[]
num=[]

for i in dataf.columns:
    if dataf[i].dtypes==object:
        cate.append(i)
    else:
        num.append(i)
dataf[num].hist(figsize=(20,12),bins=100,color="Red")
```

```python
dataf.describe(include = ['object'])
```

```python
#Visualing the categorical features with bar plots
```

```python
plt.rcParams['axes.labelsize']=20
plt.rcParams['axes.titlesize']=20
plt.rcParams['font.size']=20

fig,axes=plt.subplots(nrows=4,ncols=3,figsize=(24,18))
plt.subplots_adjust(left=None,right=None,top=None,bottom=None,wspace=None,hspace=0.4)

for i in range(len(cate)):
    dataf[cate[i]].value_counts(normalize=True).plot(kind='bar', label=cate[i], ax=axes[i//4, i%4], color='lightgreen')
    axes[i//4,i%4].set_title(cate[i])
plt.tight_layout()
```
